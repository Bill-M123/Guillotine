# Build dict for calculating player performances

Build playoff_dict with rosters, enable setting of starters

Define ownerlist (completed weeks changes every week so that is in the next cell)

```python
owners=['ls','nm','nl','pw']

playoff_dict={}
playoff_dict['rosters']={'ls':ls_players,'nm':nm_players,'nl':nl_players,'pw':pw_players,}
playoff_dict['starters']={}
```

# Set weekly starters, check legality of lineups

Currently, this block edited every week.  It needs an appropriate weekly starter dict and the completed_weeks list needs to reflect which weeks are done.

```python
completed_weeks=[15,16]

week_15_starters={'nm':['Andrews, Mark BAL TE', 'Hill, Tyreek KCC WR', 'Adams, Davante GBP WR', 'Diggs, Stefon BUF WR', 'Cook, Dalvin MIN RB', 'Brady, Tom TBB QB',
                        'Jefferson, Justin MIN WR','Mixon, Joe CIN RB'],
                 'ls':['Prescott, Dak DAL QB','Harris, Najee PIT RB','Kamara, Alvin NOS RB','Taylor, Jonathan IND RB', "Chase, Ja'Marr CIN WR",'Moore, D.J. CAR WR',
                       'Williams, Mike LAC WR', 'Gronkowski, Rob TBB TE'],
                 'pw':['Allen, Keenan LAC WR', 'Samuel, Deebo SFO WR', 'Chubb, Nick CLE RB', 'Murray, Kyler ARI QB','Edwards-Helaire, Clyde KCC RB', 'Knox, Dawson BUF TE', 
                       'Renfrow, Hunter LVR WR', 'Patterson, Cordarrelle ATL WR'],
                 'nl':['Kittle, George SFO TE', 'Kupp, Cooper LAR WR', 'Herbert, Justin LAC QB','Evans, Mike TBB WR', 'Godwin, Chris TBB WR','Ekeler, Austin LAC RB', 
                       'Jacobs, Josh LVR RB', 'Gibson, Antonio WAS RB',]}

week_16_starters={'nm':['Andrews, Mark BAL TE', 'Hill, Tyreek KCC WR', 'Adams, Davante GBP WR', 'Diggs, Stefon BUF WR', 'Cook, Dalvin MIN RB', 'Brady, Tom TBB QB',
                        'Jefferson, Justin MIN WR','Mixon, Joe CIN RB'],
                 'ls':['Stafford, Matthew LAR QB','Harris, Najee PIT RB','Kamara, Alvin NOS RB','Taylor, Jonathan IND RB', "Chase, Ja'Marr CIN WR",'Waddle, Jaylen MIA WR',
                       'Brown, Antonio TBB WR', 'Gronkowski, Rob TBB TE'],
                 'pw':['Allen, Keenan LAC WR', 'Samuel, Deebo SFO WR', 'Chubb, Nick CLE RB', 'Murray, Kyler ARI QB','Edwards-Helaire, Clyde KCC RB', 'Knox, Dawson BUF TE', 
                       'Renfrow, Hunter LVR WR', 'Patterson, Cordarrelle ATL WR'],
                 'nl':['Kittle, George SFO TE', 'Kupp, Cooper LAR WR', 'Herbert, Justin LAC QB','Evans, Mike TBB WR', 'Godwin, Chris TBB WR','Ekeler, Austin LAC RB', 
                       'Jacobs, Josh LVR RB', 'Gibson, Antonio WAS RB',]}

for week in completed_weeks:
    roster_flag=True

    weekly_starters=week_15_starters
    if week==16:
        weekly_starters=week_16_starters
    if week==17:
        weekly_starters=week_17_starters

    for key in week_15_starters:
        print(f"week{week} {key} has {len(weekly_starters[key])}starters, which are: \n{weekly_starters[key]}")
        positions=[x.split(' ')[-1] for x in weekly_starters[key]]
        rostered_positions=counter(positions)
        # Roster Checks
        if len(week_15_starters[key])>8:
            print(f"Illegal Roster:  Total Players {len(weekly_starters[key])}")
            roster_flag=False
        if rostered_positions['QB']>1:
            print(f"Illegal Roster:  Quarterbacks {rostered_positions['QB']}")
            roster_flag=False
        if rostered_positions['RB']>3:
            print(f"Illegal Roster:  Runningbacks {rostered_positions['RB']}")
            roster_flag=False
        if rostered_positions['WR']>4:
            print(f"Illegal Roster:  Wide Receivers {rostered_positions['WR']}")
            roster_flag=False
        if rostered_positions['TE']>4:
            print(f"Illegal Roster:  Tight Ends {rostered_positions['TE']}")
            roster_flag=False
        print('\n')

    if roster_flag:
        playoff_dict['starters'][f"wk_{week}"]=weekly_starters
    
```

# Develop the working dataframe

```python
playoffs_df=pd.DataFrame()

for owner in owners:

    playoffs_df[f"{owner}_roster"]=playoff_dict['rosters'][owner]
    for week in completed_weeks:
        playoffs_df[f"{owner}_wk{week}_strt"]=playoffs_df[f"{owner}_roster"].apply(lambda x: 'S' if x in playoff_dict['starters'][f"wk_{week}"][owner] else 'B')
        playoffs_df[f"{owner}_wk{week}_pts"]=0

        for starter in playoffs_df.loc[playoffs_df[f"{owner}_wk{week}_strt"]=='S',f'{owner}_roster']:
            try:
                playoffs_df.loc[playoffs_df[f"{owner}_roster"]==starter,f"{owner}_wk{week}_pts"]=player_data[player_data['Player']==starter][str(week)].values[0]
            except:
                playoffs_df.loc[playoffs_df[f"{owner}_roster"]==starter,f"{owner}_wk{week}_pts"]=0
                
playoffs_df.head(13)
    
```

```python
summary_df=pd.DataFrame(columns=['Owner','Week 15','Week 16','Week 17'])
summary_lists=[]
for owner in owners:
    tmp=[f"{owner}"]
    for week in [15,16,17]:
        try:
            tmp.append(playoffs_df[f"{owner}_wk{week}_pts"].sum())
        except:
            tmp.append(0)
    
    summary_lists.append(tmp)
summary_df=pd.DataFrame(columns=['Owner','Week 15','Week 16','Week 17'],data=summary_lists)
summary_df['Total']=summary_df['Week 15']+summary_df['Week 16']+summary_df['Week 17']
summary_df=summary_df.sort_values('Total',ascending=False)
summary_df
```
